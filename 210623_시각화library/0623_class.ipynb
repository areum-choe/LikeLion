```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

import seaborn as sns
```

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 4*np.sin(x+1.0)

plt.figure(figsize=(10,6))   # 그림의 크기
plt.plot(x, y1)
```

```python
plt.plot(x, y1, x,y2, x, y3)   # 3개의 sin 그래프
```

###tip 데이터로 데이터 살펴보고 인사이트 얻기

```python
sns.set_style("whitegrid")
```

```python
tips = sns.load_dataset("tips")  # 인터넷이 켜져 있어야 함.
tips
```

###실습1-1
 * 2개 그래프 그려보기
 * 언제 쓰이는가와 확인한 점-1줄 쓰기

```python
#요일별  tip
plt.figure(figsize=(10,7))
sns.boxplot(x="day", y="tip", data=tips)
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
# FaceGrid를 통해 그리드의 모양을 형성
g.map(sns.scatterplot, "total_bill", "tip", alpha=1.0,s=200,marker = 'X')
#scatterplot으로 안에 점을 찍음
#alpha 색 진함의 농도 설정
#marker 그래프 안에 모양
#s는 점의 크기
g.add_legend()
```
