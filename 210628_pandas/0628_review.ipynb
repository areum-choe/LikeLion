### 실습
 * sepal length, width의 행의 합을 구해보자.(sepal_sum 컬럼 생성하기)
 * petal length, width의 행의 합을 구해보자.(petal_sum 컬럼 생성하기)

```python
sepal=iris.loc[:,['sepal_length','sepal_width']]
```

```python
iris['sepal_sum']=sepal.sum(axis=1)
iris
```

```python
petal=iris.loc[:,['petal_length','petal_width']]
iris['petal_sum']=petal.sum(axis=1)
iris
```
