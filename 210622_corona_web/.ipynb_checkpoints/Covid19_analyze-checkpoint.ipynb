```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm    # 한글 폰트
import matplotlib
import os, warnings
import pandas as pd
import seaborn as sns

warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df1 = pd.read_csv('C:/Users/user/LikeLion_AI/data/CSV/today_corona_edit.csv')
df2 = pd.read_csv('C:/Users/user/LikeLion_AI/data/CSV/Natl_GDP.csv')
```

### 정확한 통계를 위해 중국의 데이터는 삭제한다...!!

```python
# df1 = df1.drop([0])

df1
```

```python
target = list(df1['국가'])
country = list(df2['국가'])

gdp = list(df2['2021'])
idx_list = []
new_cty = []
gdp_list = []

for i in target:
    for idx, one in enumerate(country):
        if i == one:
            idx_list.append(idx)
        
for i in idx_list:
    new_cty.append(country[i])
    if gdp[i] == 'no data':
        gdp_list.append(0.0)
    else:
        gdp_list.append(float(gdp[i]))

gdp_list = pd.DataFrame(gdp_list)
df1['GDP'] = gdp_list

df3 = df1.sort_values(by='인구수', ascending=False).head(30)
# df3 = df3.drop([0])
# df3 = df3.drop([1])
df3
```

```python
%matplotlib inline 
```

```python
# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## 한글 폰트 적용 ( 안하면 한글 깨짐 )
f_name = fm.FontProperties(fname="C:/Windows/Fonts/malgunbd.ttf").get_name()
plt.rc('font', family=f_name)
```

### 인도의 인구수가 다른 나라들에 비해 매우 많으므로 일시적으로 제외시켰습니다.

```python
fig = plt.figure(figsize=(10, 8))
fig.patch.set_facecolor('xkcd:white')

sns.set_color_codes('pastel')
sns.barplot(x='인구수', y='국가', data=df3, label='인구수', color='b')

sns.set_color_codes('muted')
sns.barplot(x='확진자', y='국가', data=df3, label='확진자', color='b')

fig.legend(ncol=1, loc="upper right")

sns.despine()
```

## 인구수 하위 30개국 확진자 비율

```python
df4 = df1.sort_values(by='인구수', ascending=False).tail(30)

fig = plt.figure(figsize=(10, 8))
fig.patch.set_facecolor('xkcd:white')

sns.set_color_codes('pastel')
sns.barplot(x='인구수', y='국가', data=df4, label='인구수', color='b')

sns.set_color_codes('muted')
sns.barplot(x='확진자', y='국가', data=df4, label='확진자', color='b')

fig.legend(ncol=1, loc="upper right")

sns.despine()
```

### 하위권은 상위권과 기본적으로 인구수 차이가 많이 나기때문에 보여주기용 입니다.

```python
df1['인구밀도'] = df1['인구수']/df1['국토면적']
df1
```

### < 확진자, 인구수, 국토면적, 사망자, GDP, 완치, 인구밀도 > 

```python
gdp_top10 = df1.sort_values(by='GDP', ascending=False).head(10)
gdp_low10 = df1.sort_values(by='GDP', ascending=False).tail(10)

corr_1 = gdp_top10.loc[:, ['확진자', '인구수', '국토면적', '사망자', 'GDP', '완치', '인구밀도']].corr(method='pearson')
corr_2 = gdp_low10.loc[:, ['확진자', '인구수', '국토면적', '사망자', 'GDP', '완치', '인구밀도']].corr(method='pearson')
```

```python
fig = plt.figure(figsize=(20, 8))
fig.patch.set_facecolor('xkcd:white')

axis1 = fig.add_subplot(1, 2, 1)
axis1 = sns.heatmap(corr_1, annot=True, annot_kws={'size':13}, fmt='.2f', cmap='Blues')
plt.title('< Top 10 >')
axis1.plot()

axis2 = fig.add_subplot(1, 2, 2)
axis2 = sns.heatmap(corr_2, annot=True, annot_kws={'size':13}, fmt='.2f', cmap='Blues')
plt.title('< Low 10 >')
axis2.plot()

plt.show()
```
