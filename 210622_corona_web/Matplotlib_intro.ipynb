````python
target = list(df1['국가'])
country = list(df2['국가'])

gdp = list(df2['2021'])
idx_list = []
new_cty = []
gdp_list = []

for i in target:
    for idx, one in enumerate(country): #
        if i == one:
            idx_list.append(idx)
        
for i in idx_list:
    new_cty.append(country[i])
    if gdp[i] == 'no data':
        gdp_list.append(0.0)
    else:
        gdp_list.append(float(gdp[i]))

gdp_list = pd.DataFrame(gdp_list)
df1['GDP'] = gdp_list

df1
````

````python
## 중국 데이터 삭제
# df1 = df1.drop([0])
df1['인구밀도'] = df1['인구수']/df1['국토면적']
df1_2 = df1.sort_values(by='확진자', ascending=False)
df1_2
````

````python
gdp_top10 = df1.sort_values(by='GDP', ascending=False).loc[3:14]
gdp_low10 = df1.sort_values(by='GDP', ascending=False).loc[78:89]
corr_1 = gdp_top10.loc[:, ['확진자', '인구수', '국토면적', '사망자']].corr(method='pearson')
corr_2 = gdp_low10.loc[:, ['확진자', '인구수', '국토면적', '사망자']].corr(method='pearson')
fig = plt.figure(figsize=(15, 6))
fig.patch.set_facecolor('xkcd:white')
axis1 = fig.add_subplot(1, 2, 1)
axis1 = sns.heatmap(corr_1, annot=True, annot_kws={'size':13}, fmt='.2f', cmap='Blues')
plt.title('< Top 10 >')
axis1.plot()
axis2 = fig.add_subplot(1, 2, 2)
axis2 = sns.heatmap(corr_2, annot=True, annot_kws={'size':13}, fmt='.2f', cmap='Blues')
plt.title('< Low 10 >')
axis2.plot()
````

````python
gdp_top10 = df1.sort_values(by='GDP', ascending=False).loc[3:14]
gdp_low10 = df1.sort_values(by='GDP', ascending=False).loc[78:89]
corr_1 = gdp_top10.loc[:, ['확진자', '인구수', '국토면적', '사망자','GDP','완치']].corr(method='pearson')
corr_2 = gdp_low10.loc[:, ['확진자', '인구수', '국토면적', '사망자','GDP','완치']].corr(method='pearson')
fig = plt.figure(figsize=(15, 6))
fig.patch.set_facecolor('xkcd:white')
axis1 = fig.add_subplot(1, 2, 1)
axis1 = sns.heatmap(corr_1, annot=True, annot_kws={'size':13}, fmt='.2f', cmap='Blues')
plt.title('< Top 10 >')
axis1.plot()
axis2 = fig.add_subplot(1, 2, 2)
axis2 = sns.heatmap(corr_2, annot=True, annot_kws={'size':13}, fmt='.2f', cmap='Blues')
plt.title('< Low 10 >')
axis2.plot()
````
