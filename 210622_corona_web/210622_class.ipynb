```python
sel_country = driver.find_elements_by_xpath('//*[@id="country-table"]/div/div/table/tbody/tr/td[2]')
len(sel_country)
```

```python
for one in sel_country:
    print(one.text)
```

```python
all_data = []
for i in range(2, 9):
    tmp = '//*[@id="country-table"]/div/div/table/tbody/tr/td[{}]'.format(i)
    # print(tmp)
    sel_ele = driver.find_elements_by_xpath(tmp)
    
    column_data = []
    for one in sel_ele:
        # print(one.text)
        column_data.append(one.text)
    print( len(sel_ele) )
    all_data.append(column_data)
    print(column_data)
    print()
all_data
for i in range(7):
    print(len(all_data[i]) )
```

```python
import pandas as pd
dict_dat = { "국가":all_data[0],
"확진자":all_data[1],
"사망자":all_data[2],
"완치":all_data[3],
"치명(%)":all_data[4],
"완치(%)":all_data[5],
"발생률":all_data[6] }
dat = pd.DataFrame(dict_dat)
dat
```

```python
dat.to_csv("today_corona.csv", index=False)
os.listdir( os.getcwd() )
```
