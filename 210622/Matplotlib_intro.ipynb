````python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm    # 한글 폰트
import matplotlib
import os, warnings
import pandas as pd
import seaborn as sns

warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df1 = pd.read_csv('./today_corona_edit.csv',encoding='ANSI')
df2 = pd.read_csv('./Natl_GDP.csv',encoding='ANSI')

df1
````

````python
df2[['국가', '2021']]
````

````python
target = list(df1['국가'])
country = list(df2['국가'])

gdp = list(df2['2021'])
idx_list = []
new_cty = []
gdp_list = []

for i in target:
    for idx, one in enumerate(country):
        if i == one:
            idx_list.append(idx)
        
for i in idx_list:
    new_cty.append(country[i])
    gdp_list.append(gdp[i])

gdp_list = pd.DataFrame(gdp_list)
df1['GDP'] = gdp_list

df1
````

````python
fig = plt.figure(figsize=(10, 6))
fig.patch.set_facecolor('xkcd:mint green')

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.plot(df3['인구수'], df3['확진자'])
axes2.plot(df3['확진자'], df3['치료중'])
axes3.plot(df3['확진자'], df3['완치'])
````

````python
fig = plt.figure(figsize=(10, 6))
fig.patch.set_facecolor('xkcd:mint green')

axis = fig.add_subplot(1, 1, 1)

axis.hist(df3['GDP'], bins=10)
axis.set_title('GDP')
````
