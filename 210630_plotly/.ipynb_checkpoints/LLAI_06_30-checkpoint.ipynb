### 다중 그래프 그리기
- plotly은 다음과 같음.
    - 01 make_subplots 이용하여 틀(레이아웃 만들기)
    - 02 add_trace 을 이용하여 레이아웃 위에 그래프 그리기
    - 03 update_layout을 이용하여 축제목 등을 추가하기

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True)

gap_1952 = gapminder.query("year==1952")
gap_2007 = gapminder.query("year==2007")

gap_1952_line = go.Scatter(x=gap_1952["gdpPercap"], 
                        y=gap_1952["lifeExp"], 
                        mode='markers')

gap_2007_line = go.Scatter(x=gap_2007["gdpPercap"], 
                        y=gap_2007["lifeExp"], 
                        mode='markers')

fig.add_trace(gap_1952_line, row=1, col=1)
fig.add_trace(gap_2007_line, row=2, col=1)

layout = {
    "xaxis": {
        "title": "1인당 국민소득",
    },
    "yaxis": {
        "title": "기대수명"
    },
    
    "xaxis2": {
        "title": "1인당 국민소득",
    },
    "yaxis2": {
        "title": "기대수명"
    }
}

fig.update_layout(layout)

fig
```

### 실습
- x : 인구, y : 기대수명
- 1행 2열로
- (px) 산점도 그리기

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True)

gap_1952 = gapminder.query('year==1952')
gap_2007 = gapminder.query('year==2007')

g1 = go.Scatter(x=gap_1952['pop'], y=gap_1952['lifeExp'], mode='markers')
g2 = go.Scatter(x=gap_2007['pop'], y=gap_2007['lifeExp'], mode='markers')

fig.add_trace(g1, row=1, col=1)
fig.add_trace(g2, row=2, col=1)

layout = {
    "xaxis": {
        "title": "인구수",
    },
    "yaxis": {
        "title": "기대수명"
    },
    
    "xaxis2": {
        "title": "인구수",
    },
    "yaxis2": {
        "title": "기대수명"
    }
}

fig.update_layout(layout)

fig
```

```python
fig = make_subplots(rows=2, cols=2, shared_xaxes=True)

gap_1952 = gapminder.query('year==1952')
gap_2007 = gapminder.query('year==2007')

g1 = go.Scatter(x=gap_1952['pop'], y=gap_1952['lifeExp'], 
                mode='markers', name='1952년 인구, 기대수명')
g2 = go.Scatter(x=gap_2007['pop'], y=gap_2007['lifeExp'], 
                mode='markers', name='2007년 인구, 기대수명')
g3 = go.Scatter(x=gap_1952['pop'], y=gap_1952['gdpPercap'], 
                mode='markers', name='1952년 인구, 1인당 GDP')
g4 = go.Scatter(x=gap_2007['pop'], y=gap_2007['gdpPercap'], 
                mode='markers', name='2007년 인구, 1인당 GDP')

fig.add_trace(g1, row=1, col=1)
fig.add_trace(g2, row=1, col=2)
fig.add_trace(g3, row=2, col=1)
fig.add_trace(g4, row=2, col=2)

layout = {'xaxis':{'title':'인구수'},
        'yaxis':{'title':'기대수명'},
        'xaxis2':{'title':'인구수'}, 
        'yaxis2':{'title':'기대수명'},
        'xaxis3':{'title':'인구수'},
        'yaxis3':{'title':'1인당 국민소득'},
        'xaxis4':{'title':'인구수'}, 
        'yaxis4':{'title':'1인당 국민소득'}
        }

fig.update_layout(layout)

fig.show()
```
